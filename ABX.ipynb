```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)

data = "/content/drive/MyDrive/solicitacoescredito.csv"
df = pd.read_csv(data)
df.head(10)
```

```python
df.info()
```

```python
## LIMPEZA DE DADOS

# Tratamento de valores ausentes para variáveis financeiras
col_financeira = ['ativoCirculante', 'passivoCirculante', 'totalAtivo', 'totalPatrimonioLiquido',
                  'endividamento', 'duplicatasAReceber', 'estoque', 'faturamentoBruto', 'margemBruta',
                  'custos', 'limiteEmpresaAnaliseCredito','capitalSocial']

for col in col_financeira:
    df[col].fillna(df[col].median(), inplace=True)


# Tratamento de valores ausentes para outras variáveis numéricas
col_numerica = ['percentualProtestos', 'valorAprovado', 'diferencaPercentualRisco', 'percentualRisco',
                'dashboardCorrelacao', 'scorePontualidade','periodoDemonstrativoEmMeses','anoFundacao' ]

for col in col_numerica:
    df[col].fillna(df[col].mean(), inplace=True)


# Tratamento de valores ausentes para variáveis categóricas e datas
col_categorica = ['primeiraCompra', 'dataAprovadoEmComite', 'periodoBalanco', 'intervaloFundacao',
                    'restricoes', 'empresa_MeEppMei', 'dataAprovadoNivelAnalista']

for col in col_categorica:
    df[col].fillna(df[col].mode()[0], inplace=True)

```
